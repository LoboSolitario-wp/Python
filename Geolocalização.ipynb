```python
import geopandas as gpds
import pandas as pd
```

```python
gpds.tools.geocode("Umuarama, São paulo", provider='nominatim', user_agent="my-application")
```

```python
df = pd.read_excel('Arquivo.xlsx')
df.head()
```

```python
gpds.tools.geocode(df['Orig real'], provider='nominatim', user_agent="my-application")
gpds.tools.geocode(df['Dest real'], provider='nominatim', user_agent="my-application")
```

```python
df['CoordOrigem'] = gpds.tools.geocode(df['Orig real'], provider='nominatim', user_agent="my-application")['geometry']
df['CoordDestin'] = gpds.tools.geocode(df['Dest real'], provider='nominatim', user_agent="my-application")['geometry']
```
